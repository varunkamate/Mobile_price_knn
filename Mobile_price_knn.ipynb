```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv("train.csv")
df1=pd.read_csv("test.csv")
```

```python
#colummn price_range present in the df is the output for both data we have to build the model to predict the output
#check the null values in data
# it doesn't ccontainn any null value in it so  their is no need of the feature extraction
df.isna().sum()
```

```python
df.columns
```

```python
#For train the data set we have to split the data into 2 parts one fro input and other one for output
#for input
x=df[['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
       'touch_screen', 'wifi']]
```

```python
#for output
y=df[['price_range']]
```

```python
#for model building we use the trrain_test_split  for to define train and test data for both input and output
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
x_train
```

```python
#creating model
model=KNeighborsClassifier()
```

```python
#model calling
model.fit(x_train,y_train)
```

```python
#have to define the prediction variable
y_pred=model.predict(x_test)
```

```python
#for the accuracy finding we have to print the classification report
print(classification_report(y_test,y_pred))
```
